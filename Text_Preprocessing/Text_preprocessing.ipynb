```python
import string
punc = string.punctuation
sentence = "The United States Internation University is an awsome place ! I love # USIU"
without_punc = [word for word in sentence.split() if word not in punc]
print("Sentence without Punctuations:", " ".join(without_punc))
```

Removing Special Characters

```python
import re
sentence = "Find the remainder when [math]23^{24}[/math] is divided by 24,23?"
sentence_clean = re.sub("[^a-zA-Z\s]", " ", sentence)
print("Clean Sentence:", sentence_clean)
```
