```python
#1. import libraries
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
# Install gensim if not installed
!pip install gensim
from gensim import corpora, models
from gensim.models import LdaModel
from gensim.corpora import Dictionary
# Install pyLDAvis if not installed
!pip install pyLDAvis
import pyLDAvis.gensim_models as gensimvis
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
#2.
documents = [
    "The customer completed a payment of 500 KES via mobile money",
    "A suspicious transfer was completed from account 12345 to an unknown merchant",
    "The user loaged in from a new device located in Nairobi.", "Multiple failed login attempts were recorded in the last hour.",
    "A refund reques submitted for a transaction that never occurred.",
    "The account balance dropped sharply after an unusually large withdrawal.",
    "An alert was triggered due to high-frequency transactions in a short time frame."
]
```

```python
#3.
stopwords = set(stopwords.words('english'))

def preprocess(text):
  tokens = word_tokenize(text.lower())
  filtered_tokens = [word for word in tokens if word.isalpha() and word not in stopwords]
  return filtered_tokens

processed_docs = [preprocess(doc) for doc in documents]
```

```python
#4. Create a Dictionary and Corpus
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
#5. Train the LDA model
NUM_TOPICS = 3
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=NUM_TOPICS, passes = 10, iterations = 100, random_state = 42)
```

```python
#6. View Topics
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx + 1}:{topic}")
```
